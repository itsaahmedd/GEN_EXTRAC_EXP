```python
#!/usr/bin/env python
import argparse
import json
import torch
from datasets import Dataset
import os
from transformers import (
    AutoTokenizer,
    AutoModelForQuestionAnswering,
)
from torch.utils.data import DataLoader
import numpy as np
from rouge import Rouge
import csv
import time
import matplotlib.pyplot as plt
from tqdm import tqdm  # Import tqdm for progress bars

import evaluate  # Hugging Face evaluate

from sentence_transformers import SentenceTransformer, util
from sklearn.metrics import precision_recall_curve, average_precision_score

file_path = "/content/drive/My Drive/Dissertation/cuad_qa_dataset.json"


def read_json(file_path):

    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    return data



def data_preparation(data):
    # We'll just take a small subset (5430) for demonstration
    raw_data = data["test"][:2000]


    # Convert dataset to a huggingface Dataset
    dataset = Dataset.from_list(raw_data)
    print(dataset)

    print("Test Dataset Size: ", len(dataset))


    return dataset


def prepare_test_features(entries, tokenizer, max_length=512, doc_stride=128):

    """
    Tokenizes a batch of examples for evaluation.
    Returns a dict with keys: input_ids, attention_mask, offset_mapping, and a mapping of each chunk to its example.
    Assumes each example has at least "question", "context", and a unique "id" field.
    """
    encodings = tokenizer(
        entries["question"],
        entries["context"],
        truncation="only_second",
        max_length=max_length,
        stride=doc_stride,
        return_overflowing_tokens=True,
        return_offsets_mapping=True,
        padding="max_length",
    )

    # This mapping tells us which generated chunk belongs to which original example.
    sample_mapping = encodings.pop("overflow_to_sample_mapping")

    # Attach the mapping so we can later combine chunks.
    encodings["entry_id"] = [entries["id"][i] for i in sample_mapping]

    return encodings

def postprocess_predictions(test_examples, features, all_start_logits, all_end_logits, tokenizer, max_answer_length=30):
    """
    For each example, combine the predictions from all its chunks.
    This simplified version scores each chunk's candidate answer span using the sum of start and end logits,
    then picks the answer with the highest score.
    """
    # Build a mapping from entry_id to list of (feature_idx, offsets, input_ids, start_logits, end_logits)
    entry_to_features = {}

    for idx, en_id in enumerate(features["entry_id"]):

        if en_id not in entry_to_features:
            entry_to_features[en_id] = []
        entry_to_features[en_id].append(idx)

    predictions = {}

    for entry in test_examples:

        en_id = entry["id"]
        best_score = -1e9
        best_answer = ""

        if en_id not in entry_to_features:
            predictions[en_id] = ""
            continue

        for feat_idx in entry_to_features[en_id]:
            # Get logits and offset mapping for this feature.

            start_logits = all_start_logits[feat_idx]
            end_logits = all_end_logits[feat_idx]
            offsets = features["offset_mapping"][feat_idx]
            input_ids = features["input_ids"][feat_idx]

            # Loop over all possible spans.
            for i in range(len(start_logits)):
                # Skip if the offset is (0,0): these are special tokens.
                if offsets[i][0] == offsets[i][1]:
                    continue

                for j in range(i, min(i + max_answer_length, len(end_logits))):
                    if offsets[j][0] == offsets[j][1]:
                        continue

                    # Only consider spans that are in the context.
                    if offsets[i][0] > 0 and offsets[j][1] > 0:
                        score = start_logits[i] + end_logits[j]
                        if score > best_score:
                            best_score = score
                            best_answer = tokenizer.decode(input_ids[i:j+1], skip_special_tokens=True)

        predictions[en_id] = best_answer

    return predictions



def compute_f1(prediction, ground_truth):
    def tokenize(text):
        return text.split()
    pred_tokens = tokenize(prediction)
    gt_tokens = tokenize(ground_truth)
    if len(pred_tokens) == 0 and len(gt_tokens) == 0:
        return 1.0
    common = set(pred_tokens) & set(gt_tokens)
    if len(common) == 0:
        return 0.0
    precision = len(common) / len(pred_tokens)
    recall = len(common) / len(gt_tokens)
    return 2 * (precision * recall) / (precision + recall)


def evaluate_predictions(predictions, test_examples, sbert_model, rouge_metric, debug=False):
    results_details = []  # per-example details
    labels = []
    confidences = []
    f1_scores = []
    semantic_scores = []
    rouge_scores = []
    times = []  # time per example

    squad_predictions = {}
    squad_references = {}

    for entry in test_examples:
        ex_id = entry["id"]
        question = entry["question"]
        context = entry["context"]
        ref = entry["answers"]
        gold_answers = ref["text"]
        if not gold_answers or len(gold_answers) == 0:
            ref["text"] = [""]
            ref["answer_start"] = [0]
        gold_text = ref["text"][0].strip()
        pred_text = predictions.get(ex_id, "").strip()

        start_time = time.time()
        f1 = compute_f1(pred_text, gold_text)
        duration = time.time() - start_time
        times.append(duration)

        f1_scores.append(f1)
        label = 1 if f1 >= 0.6 else 0
        labels.append(label)
        confidences.append(f1)

        if gold_text == "" or pred_text == "":
            sem_sim = 0.0
        else:
            emb_gold = sbert_model.encode(gold_text, convert_to_tensor=True)
            emb_pred = sbert_model.encode(pred_text, convert_to_tensor=True)
            sem_sim = float(util.pytorch_cos_sim(emb_gold, emb_pred).item())
        semantic_scores.append(sem_sim)

        if gold_text == "" or pred_text == "":
            r_score = 0.0
        else:
            r = rouge_metric.get_scores(pred_text, gold_text)
            r_score = r[0]["rouge-l"]["f"]
        rouge_scores.append(r_score)

        results_details.append({
            "id": ex_id,
            "question": question,
            "context": context,
            "gold_answer": gold_text,
            "model_answer": pred_text,
            "f1": f1,
            "semantic_similarity": sem_sim,
            "rouge_l_f1": r_score,
            "eval_time_sec": duration
        })

        squad_predictions[ex_id] = pred_text
        squad_references[ex_id] = {"id": ex_id, "answers": ref}

        if debug:
            print("------------------------------------------------------------")
            print(f"ID: {ex_id}")
            print(f"Question: {question}")
            # Optionally, print context if needed
            # print(f"Context: {context}")
            print(f"Gold Answer: {gold_text}")
            print(f"Model Answer: {pred_text}")
            print("------------------------------------------------------------\n")

    squad_preds = [{"id": ex_id, "prediction_text": pred_text} for ex_id, pred_text in squad_predictions.items()]
    squad_refs = list(squad_references.values())

    precision_vals, recall_vals, _ = precision_recall_curve(labels, confidences)
    ap_score = average_precision_score(labels, confidences)
    p_at_80 = max([p for p, r in zip(precision_vals, recall_vals) if r >= 0.8] or [0.0])

    squad_metric = evaluate.load("squad")
    squad_results = squad_metric.compute(predictions=squad_preds, references=squad_refs)

    overall_results = {
        "avg_f1": float(np.mean(f1_scores)),
        "avg_semantic_similarity": float(np.mean(semantic_scores)),
        "avg_rouge_l_f1": float(np.mean(rouge_scores)),
        "AUPR": float(ap_score),
        "precision_at_80%": float(p_at_80),
        "SQuAD_EM": squad_results.get("exact_match", None),
        "SQuAD_F1": squad_results.get("f1", None),
        "avg_eval_time_sec": float(np.mean(times))
    }

    return overall_results, results_details


def save_results_to_csv(details, filename):
    keys = details[0].keys()
    with open(filename, "w", newline="") as f:
        dict_writer = csv.DictWriter(f, keys)
        dict_writer.writeheader()
        dict_writer.writerows(details)


def plot_model_comparison(aggregated_results):
    # aggregated_results is a list of dictionaries, one per model, with a "model" key.
    models = [r["model"] for r in aggregated_results]
    avg_f1 = [r["avg_f1"] for r in aggregated_results]
    avg_rouge = [r["avg_rouge_l_f1"] for r in aggregated_results]
    avg_semantic = [r["avg_semantic_similarity"] for r in aggregated_results]

    x = np.arange(len(models))
    width = 0.25

    plt.figure(figsize=(10, 6))
    plt.bar(x - width, avg_f1, width, label="F1")
    plt.bar(x, avg_rouge, width, label="ROUGE-L")
    plt.bar(x + width, avg_semantic, width, label="Semantic Similarity")
    plt.xlabel("Model")
    plt.ylabel("Metric Score")
    plt.title("Evaluation Metrics Comparison Across Models")
    plt.xticks(x, models, rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

def main():
    raw_data = read_json(file_path)
    test_examples = data_preparation(raw_data)

    # We'll aggregate overall results for each model in a list.
    aggregated_results = []
    all_details = {}

    base_model_dir = "/content/drive/My Drive/Dissertation/model_output"
    batch_size = 16

    sbert_model = SentenceTransformer("all-MiniLM-L6-v2")
    rouge_metric = Rouge()

    for model_subdir in tqdm(os.listdir(base_model_dir), desc="Evaluating models"):
        model_dir = os.path.join(base_model_dir, model_subdir)
        if not os.path.isdir(model_dir):
            continue

        print(f"\n==== Evaluating model from: {model_dir} ====\n")
        tokenizer = AutoTokenizer.from_pretrained(model_dir, use_fast=True)
        model = AutoModelForQuestionAnswering.from_pretrained(model_dir)
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        model.to(device)
        model.eval()

        # Remove all original columns for tokenization.
        columns_to_remove = test_examples.column_names
        tokenised_test = test_examples.map(
            lambda examples: prepare_test_features(examples, tokenizer),
            batched=True,
            remove_columns=columns_to_remove
        )

        test_dataloader = DataLoader(tokenised_test, batch_size=batch_size)
        all_start_logits = []
        all_end_logits = []
        features = {
            "input_ids": tokenised_test["input_ids"],
            "offset_mapping": tokenised_test["offset_mapping"],
            "entry_id": tokenised_test["entry_id"],
        }

        with torch.no_grad():
          for batch in tqdm(test_dataloader, desc=f"Processing batches for {os.path.basename(model_dir)}"):
                input_ids = torch.stack([torch.tensor(x) for x in batch["input_ids"]]).to(device)
                attention_mask = torch.stack([torch.tensor(x) for x in batch["attention_mask"]]).to(device)
                outputs = model(input_ids=input_ids, attention_mask=attention_mask)
                all_start_logits.extend(outputs.start_logits.cpu().numpy())
                all_end_logits.extend(outputs.end_logits.cpu().numpy())

        predictions = postprocess_predictions(test_examples, features, all_start_logits, all_end_logits, tokenizer)
        overall_results, details = evaluate_predictions(predictions, test_examples, sbert_model, rouge_metric, debug=False)

        # Save details for this model (use model directory name as identifier)
        model_name = os.path.basename(model_dir)
        overall_results["model"] = model_name
        aggregated_results.append(overall_results)
        all_details[model_name] = details

        # Optionally, save per-model CSV:
        csv_filename = f"evaluation_details_{model_name}.csv"
        save_results_to_csv(details, csv_filename)
        print(f"Saved details to {csv_filename}")
        print("Overall Results:")
        print(overall_results)

    # Save aggregated results (across models) to a CSV.
    aggregated_csv = "aggregated_evaluation_results.csv"
    with open(aggregated_csv, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=aggregated_results[0].keys())
        writer.writeheader()
        writer.writerows(aggregated_results)
    print(f"Saved aggregated results to {aggregated_csv}")

    # Plot comparison graphs.
    plot_model_comparison(aggregated_results)


if __name__ == "__main__":
    main()
```

```python
import csv
from google.colab import files


def load_aggregated_results(csv_filename):
    with open(csv_filename, "r") as f:
        reader = csv.DictReader(f)
        results = [row for row in reader]
    # Convert numeric values from strings to floats if needed:
    for r in results:
        r["avg_f1"] = float(r["avg_f1"])
        r["avg_rouge_l_f1"] = float(r["avg_rouge_l_f1"])
        r["avg_semantic_similarity"] = float(r["avg_semantic_similarity"])
        r["avg_eval_time_sec"] = float(r["avg_eval_time_sec"])
    return results






def plot_model_comparison(aggregated_results):
    models = [r["model"] for r in aggregated_results]
    avg_f1 = [r["avg_f1"] for r in aggregated_results]
    avg_rouge = [r["avg_rouge_l_f1"] for r in aggregated_results]
    avg_semantic = [r["avg_semantic_similarity"] for r in aggregated_results]
    avg_time = [r.get("avg_eval_time_sec", 0) for r in aggregated_results]

    x = np.arange(len(models))
    width = 0.25

    # Plot evaluation metrics comparison
    plt.figure(figsize=(10, 6))
    plt.bar(x - width, avg_f1, width, label="F1")
    plt.bar(x, avg_rouge, width, label="ROUGE-L")
    plt.bar(x + width, avg_semantic, width, label="Semantic Similarity")
    plt.xlabel("Model")
    plt.ylabel("Metric Score")
    plt.title("Evaluation Metrics Comparison Across Models")
    plt.xticks(x, models, rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.savefig("metrics_comparison.png")  # Save plot as file
    plt.show()

    # Plot evaluation time comparison
    plt.figure(figsize=(10, 6))
    plt.bar(x, avg_time, width=0.5, color="skyblue")
    plt.xlabel("Model")
    plt.ylabel("Average Evaluation Time (sec)")
    plt.title("Average Evaluation Time per Example Across Models")
    plt.xticks(x, models, rotation=45)
    plt.tight_layout()
    plt.savefig("avg_eval_time.png")  # Save plot as file
    plt.show()


aggregated_results = load_aggregated_results("aggregated_evaluation_results.csv")
plot_model_comparison(aggregated_results)

# files.download("metrics_comparison.png")

```
